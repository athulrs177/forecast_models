```python
def process_year(test_year, gamma_cnn):
# for t, test_year in enumerate(years):    
    print('test_year: '+str(test_year))
    ############################################################################
    # load the test and train files and for the appropriate season
    t_train_imerg = pd.to_datetime([])
    t_test_imerg = pd.to_datetime([])
    
    t_train_waves = pd.to_datetime([])
    t_test_waves = pd.to_datetime([])

    fili_train = f'*imerg_amplitude_phase_theta.{accum_days}days.train.{test_year}.nc'
    fili_test = f'*imerg_amplitude_phase_theta.{accum_days}days.test.{test_year}.nc'

    # train-dates
    train_years = years[years!=test_year]
    for train_year in train_years:
        start_date_train=pd.to_datetime(strt+str(train_year), format='%m/%d/%Y')
        end_date_train=pd.to_datetime(last+str(train_year), format='%m/%d/%Y')
        
        t0_imerg = pd.date_range(start=start_date_train, end=end_date_train)
        
        t0_waves = pd.date_range(start=strtPred+str(train_year), end=lastPred+str(train_year))
        
        if calendar.isleap(train_year):
            t0_imerg = t0_imerg.drop(pd.Timestamp(train_year, 2, 29), errors='ignore')
            t0_waves = t0_waves.drop(pd.Timestamp(train_year, 2, 29), errors='ignore')
        t_train_imerg  = t_train_imerg.union(t0_imerg)
        t_train_waves  = t_train_waves.union(t0_waves)

    # test-dates
    
    start_date_test=pd.to_datetime(strt+str(test_year), format='%m/%d/%Y')
    end_date_test=pd.to_datetime(last+str(test_year), format='%m/%d/%Y') 
    
    t00_imerg = pd.date_range(start=start_date_test, end=end_date_test)
    
    t00_waves = pd.date_range(start=strtPred+str(test_year), end=lastPred+str(test_year))
    
    if calendar.isleap(test_year):
        t00_imerg = t00_imerg.drop(pd.Timestamp(test_year, 2, 29), errors='ignore')
        t00_waves = t00_waves.drop(pd.Timestamp(test_year, 2, 29), errors='ignore')
    t_test_imerg  = t_test_imerg.union(t00_imerg)
    t_test_waves  = t_test_waves.union(t00_waves)
    
    #################################################
    # Lat0 Lon0
    #################################################
    f_train = xr.open_mfdataset(diri+fili_train, combine='nested', concat_dim='waves', compat='override', coords='minimal').sel(time=t_train_waves)
    f_train = f_train.assign_coords({'waves': wave_id})
    
    f_test = xr.open_mfdataset(diri+fili_test, combine='nested', concat_dim='waves', compat='override', coords='minimal').sel(time=t_test_waves)
    f_test = f_test.assign_coords({'waves': wave_id})
    
    
    f_train_tar = f_train.sel(lat=lati, lon=loni)
    f_test_tar = f_test.sel(lat=lati, lon=loni)
    
    ##################################################
    # LatPlus0 LonPlus3
    ##################################################
    f_train_latPlus0_lonPlus3 = merge_adjusted_datasets(f=f_train, up_down='up',  wave_id_after=wave_id_latPlus0_lonPlus3, 
                                                        lati=lati, loni=loni, latAdjustment=0, lon_step=3)
    
    f_test_latPlus0_lonPlus3 = merge_adjusted_datasets(f=f_test, up_down='up', wave_id_after=wave_id_latPlus0_lonPlus3, 
                                                       lati=lati, loni=loni, latAdjustment=0, lon_step=3)
    
    ##################################################
    # LatPlus0 LonMinus3
    ##################################################
    f_train_latPlus0_lonMinus3 = merge_adjusted_datasets(f=f_train, up_down='down', wave_id_after=wave_id_latPlus0_lonMinus3, 
                                                        lati=lati, loni=loni, latAdjustment=0, lon_step=3)
    
    f_test_latPlus0_lonMinus3 = merge_adjusted_datasets(f=f_test, up_down='down', wave_id_after=wave_id_latPlus0_lonMinus3, 
                                                       lati=lati, loni=loni, latAdjustment=0, lon_step=3)
    
    ##################################################
    # LatPlus0 LonPlus5
    ##################################################
    f_train_latPlus0_lonPlus5 = merge_adjusted_datasets(f=f_train, up_down='up',  wave_id_after=wave_id_latPlus0_lonPlus5, 
                                                        lati=lati, loni=loni, latAdjustment=0, lon_step=5)
    
    f_test_latPlus0_lonPlus5 = merge_adjusted_datasets(f=f_test, up_down='up', wave_id_after=wave_id_latPlus0_lonPlus5, 
                                                       lati=lati, loni=loni, latAdjustment=0, lon_step=5)
    
    ##################################################
    # LatPlus0 LonMinus5
    ##################################################
    f_train_latPlus0_lonMinus5 = merge_adjusted_datasets(f=f_train, up_down='down', wave_id_after=wave_id_latPlus0_lonMinus5, 
                                                        lati=lati, loni=loni, latAdjustment=0, lon_step=5)
    
    f_test_latPlus0_lonMinus5 = merge_adjusted_datasets(f=f_test, up_down='down', wave_id_after=wave_id_latPlus0_lonMinus5, 
                                                       lati=lati, loni=loni, latAdjustment=0, lon_step=5)
    
    ##################################################
    # LatPlus0 LonPlus7
    ##################################################
    f_train_latPlus0_lonPlus7 = merge_adjusted_datasets(f=f_train, up_down='up',  wave_id_after=wave_id_latPlus0_lonPlus7, 
                                                        lati=lati, loni=loni, latAdjustment=0, lon_step=7)
    
    f_test_latPlus0_lonPlus7 = merge_adjusted_datasets(f=f_test, up_down='up', wave_id_after=wave_id_latPlus0_lonPlus7, 
                                                       lati=lati, loni=loni, latAdjustment=0, lon_step=7)
    
    ##################################################
    # LatPlus0 LonMinus7
    ##################################################
    f_train_latPlus0_lonMinus7 = merge_adjusted_datasets(f=f_train, up_down='down', wave_id_after=wave_id_latPlus0_lonMinus7, 
                                                        lati=lati, loni=loni, latAdjustment=0, lon_step=7)
    
    f_test_latPlus0_lonMinus7 = merge_adjusted_datasets(f=f_test, up_down='down', wave_id_after=wave_id_latPlus0_lonMinus7, 
                                                       lati=lati, loni=loni, latAdjustment=0, lon_step=7)
    
    ##################################################
    # LatPlus0 LonPlus9
    ##################################################
    f_train_latPlus0_lonPlus9 = merge_adjusted_datasets(f=f_train, up_down='up',  wave_id_after=wave_id_latPlus0_lonPlus9, 
                                                        lati=lati, loni=loni, latAdjustment=0, lon_step=9)
    
    f_test_latPlus0_lonPlus9 = merge_adjusted_datasets(f=f_test, up_down='up', wave_id_after=wave_id_latPlus0_lonPlus9, 
                                                       lati=lati, loni=loni, latAdjustment=0, lon_step=9)
    
    ##################################################
    # LatPlus0 LonMinus9
    ##################################################
    f_train_latPlus0_lonMinus9 = merge_adjusted_datasets(f=f_train, up_down='down', wave_id_after=wave_id_latPlus0_lonMinus9, 
                                                        lati=lati, loni=loni, latAdjustment=0, lon_step=9)
    
    f_test_latPlus0_lonMinus9 = merge_adjusted_datasets(f=f_test, up_down='down', wave_id_after=wave_id_latPlus0_lonMinus9, 
                                                       lati=lati, loni=loni, latAdjustment=0, lon_step=9)
    
    ##################################################
    # Final
    ##################################################    
    
    f_train_final = xr.concat([f_train_tar, 
                               # f_train_latPlus0_lonPlus1, f_train_latPlus0_lonMinus1,
                               f_train_latPlus0_lonPlus3, f_train_latPlus0_lonMinus3, 
                               f_train_latPlus0_lonPlus5, f_train_latPlus0_lonMinus5,
                               f_train_latPlus0_lonPlus7, f_train_latPlus0_lonMinus7, 
                               f_train_latPlus0_lonPlus9, f_train_latPlus0_lonMinus9], dim='waves')    
    f_test_final = xr.concat([f_test_tar, 
                              # f_test_latPlus0_lonPlus1, f_test_latPlus0_lonMinus1, 
                              f_test_latPlus0_lonPlus3, f_test_latPlus0_lonMinus3, 
                              f_test_latPlus0_lonPlus5, f_test_latPlus0_lonMinus5,
                              f_test_latPlus0_lonPlus7, f_test_latPlus0_lonMinus7, 
                              f_test_latPlus0_lonPlus9, f_test_latPlus0_lonMinus9], dim='waves')
    
    

    # Transform amplitude based on phase values
    phase_adjusted_amplitude_train = f_train_final.amplitude * np.cos( np.deg2rad(f_train_final.theta) )
    # Assign phase_adjusted_amplitude to waves1 dataset
    waves_train = phase_adjusted_amplitude_train
    
    # Transform amplitude based on phase values
    phase_adjusted_amplitude_test = f_test_final.amplitude * np.cos( np.deg2rad(f_test_final.theta) )
    # Assign phase_adjusted_amplitude to waves1 dataset
    waves_test = phase_adjusted_amplitude_test

    imerg_train = imerg.sel(time=t_train_imerg.floor('D'))
    imerg_test  = imerg.sel(time=t_test_imerg.floor('D'))
    waves_train = waves_train.transpose('time', 'waves')
    waves_test  = waves_test.transpose('time', 'waves')
    
    forecast_model = ForecastModel(params=params)
    
    waves_in = (waves_train, waves_test)
    imerg_in = (imerg_train, imerg_test)
    
    if gamma_cnn == 'cnn':
        history = forecast_model.cnn_fit_predict(waves=waves_in, 
                                                 imerg=imerg_in, 
                                                 train_years=train_years
                                                )
    elif gamma_cnn == 'gamma':
        history = forecast_model.GammaRegressor_fit_predict(waves=waves_in, 
                                                            imerg=imerg_in, 
                                                            train_years=train_years
                                                           )
    
    else:
        raise(f'{gamma_cnn} model is not supported yet!')
    
    
    results = {'predictions': history['predictions'],
               'feature_importance': history['feature_importance'],
               'selected_features_mask': history['selected_features_mask'],
               'selected_features': history['selected_features'],
               'top_features': history['top_features'],
               'n_features_selected': history['n_features_selected'],
               'waves_train': history['waves_train']
              }

    return results
    # break
```

```python
%%time
results = Parallel(n_jobs=13, verbose=10)(delayed(process_year)(test_year, gamma_cnn) for test_year in years if test_year >= 2007)
imerg_pred = np.concatenate(np.array([res['predictions'] for res in results]), axis=0)
feature_importance = np.array([res['feature_importance'] for res in results])
# selected_features = np.array([res['selected_features'] for res in results])
top_features = np.array([res['top_features'] for res in results])
n_features_selected = np.array([res['n_features_selected'] for res in results])
selected_features_mask = np.array([res['selected_features_mask'] for res in results])
waves_train = np.concatenate(np.array([res['waves_train'] for res in results]), axis=0)
```

```python
def taylor_skill_score(forecast_variance, correlation, max_correlation = 1.0):

    score = 4*(1+correlation) /( (forecast_variance + (1/forecast_variance) )**2 * ( 1 + max_correlation ) )
    return score
```
